```python
import matplotlib.pyplot as plt
import geopandas as gpd
```

### Explore visualization tools from matplotlib

### Shapefile from https://www.census.gov/geographies/mapping-files/time-series/geo/carto-boundary-file.html

```python
county_data = gpd.read_file("cb_2018_us_county_500k.zip")
state_data = gpd.read_file("cb_2018_us_state_500k.zip")
```

```python
intersected_county = county_data.to_crs(state_data.crs)
intersected_county = gpd.overlay(intersected_county, 
                                 state_data, 
                                 how='intersection',
                                keep_geom_type=True
                                )
intersected_county['area'] = intersected_county.to_crs("EPSG:5070").area
```

### Test out different format from https://matplotlib.org/stable/gallery/style_sheets/

```python
plt.style.use("dark_background")
ax = state_data.boundary.plot(facecolor=None, linewidth=0.5, figsize=(5,7),
                              color="white"
                             )
intersected_county.plot(
    "area",
    ax=ax, 
    linewidth=0.3,
    )
plt.tight_layout()
```

```python
plt.style.use('Solarize_Light2')
ax = state_data.boundary.plot(facecolor=None, linewidth=0.5, figsize=(5,7),
                              color="black"
                             )
intersected_county.plot(
    "area",
    ax=ax, 
    linewidth=0.3,
    )
plt.tight_layout()
```

```python
plt.style.use("bmh")
ax = state_data.boundary.plot(facecolor=None, linewidth=0.5, figsize=(5,7),
                              color="black"
                             )
intersected_county.plot(
    "area",
    ax=ax, 
    linewidth=0.3,
    )
plt.tight_layout()
```
